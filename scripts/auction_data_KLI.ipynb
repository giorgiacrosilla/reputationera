initial search on auctions data. 

```python
import pandas as pd
pd.set_option('display.max_columns', None)
getty2 = pd.read_csv("auctions.csv", sep=";", encoding="iso-8859-1")

getty2
```

```python
from collections import Counter
liss = {}
for column_name in getty2.columns:
    if column_name == "artist":
        for value in getty2[column_name]:
            if value not in liss:
                liss[value] = 0
            else:
                liss[value] += 1
k = Counter(liss)
 
# Finding 3 highest values
high = k.most_common(4) 
 
print("Initial Dictionary:")
print(liss, "\n")
 
 
print("Dictionary with 3 highest values:")
print("Keys: Values")
 
for i in high:
    print(i[0]," :",i[1]," ")
```

```python
liss = []
for column_name in getty2.columns:
    if column_name == "artist":
        for value in getty2[column_name]:
            if value == "Gustav Klimt":
                liss.append(value)
print(len(liss))
```

Get only results related to sales on Klimt works. Artworks that do not possess data about price or soldtime have been deleted from the dataset.
The dataset does not possess any data related to sellers and buyers. 
In this example we notice that 256 works over a total of 424 possess data about price and time of the transaction. 

```python
import pandas as pd

new_dataframe = getty2[["artist", "name", "price", "soldtime"]].copy()
new_dataframe = new_dataframe[new_dataframe["artist"] == "Gustav Klimt"].copy()

new_dataframe.reset_index(drop=True, inplace=True)

for idx, row in new_dataframe.iterrows():
    if pd.isna(row["price"]) and pd.isna(row["soldtime"]):
        new_dataframe.drop(idx, inplace=True)
        
dayy = []
for i in new_dataframe['soldtime']:
    string_representation = str(i)
    split_parts = string_representation.split('-')
    day = split_parts[0]
    dayy.append(day)
new_dataframe['soldtime'] = dayy
        

new_dataframe.reset_index(drop=True, inplace=True)
new_dataframe
```
